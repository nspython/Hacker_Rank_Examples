Write a NumPy program to compute the x and y coordinates for points on a sine curve and plot the points using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4 * np.pi, 0.4)
y = np.sin(x)
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.show()
```
